```python
import matplotlib.pyplot as plt
from importlib import reload
import pandas as pd
import numpy as np
import autoencoder 
reload(autoencoder)

import seaborn as sns
from pylab import rcParams

sns.set(style='whitegrid', palette='muted', font_scale=1.5)
rcParams['figure.figsize'] = 14, 8

RANDOM_SEED = 42

%matplotlib inline
```

```python
df = pd.read_csv('feature_data.csv')
df.shape
```

```python
X = df.drop(['id','event_id','timein','timeout'],axis=1)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_X = pca.fit_transform(X)
```

```python
pca_X
```

```python
dims = (pca_X.shape[1], 8)
AE = autoencoder.AutoencoderNN(dims=dims, batch_size=30, nb_epoch=100)
```

```python
model, history = AE.train(pca_X)
```

```python
pred = AE.predict(model, pca_X)
```

## Reconstruction Error

```python
mse = np.mean(np.power(pca_X - pred, 2), axis=1)
error_df = pd.DataFrame({'reconstruction_error': mse})
error_df.head()
```

```python
error_df['anomalous'] = 0
error_df.loc[error_df.reconstruction_error > 10000, 'anomalous'] = 1
```

```python
threshold = 10000

fig, ax = plt.subplots(figsize=(14,8))
ax.plot(error_df.index, error_df.reconstruction_error, marker='o', ms=3.5, linestyle='')

ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Reconstruction error for different classes")
plt.ylabel("Reconstruction error")
plt.xlabel("Data point index")
plt.show();
```

```python
error_df.anomalous.value_counts()
```

```python
top_20_df = df.loc[df['anomalous'] == 1, 'id'].value_counts()[:20]
top_20_df.plot(kind='barh', figsize=(10,12))
plt.xticks(range(20))
plt.title('Top 20 passengers with the highest flags')
plt.xlabel('# of flags')
plt.ylabel("'Passenger ID's")
plt.show()
```

## Network Analysis
